```python
# check the missing value
df.isna().sum()
```

```python
# Find rows with any missing value
rows_with_missing_data = df[df.isnull().any(axis=1)]

print("\nRows with missing values:")
print(rows_with_missing_data)
```

```python
# Check the total percentage of missing data in  model column
missing_percentage_model = df['model'].isnull().mean() * 100
print(f"Percentage missing in model: {missing_percentage_model}%")

# Check the total percentage of missing data in sentiment column
missing_percentage_senti = df['sentiment'].isnull().mean() * 100
print(f"Percentage missing in sentiment: {missing_percentage_senti}%")
```

```python

counts_senti = df['sentiment'].value_counts()
counts_model = df['model'].value_counts()
print(counts_senti)
print(counts_model)
```

```python
df_clean = df.dropna(subset=['model','sentiment'],how='any')
df_clean.head()
```

```python
df_clean.isna().sum()
```

```python
df_clean.info()
```

```python

import re
# Option 1: Explicitly targeting NBSP (\u00A0) and standard whitespace
df_clean['review'] = df_clean['review'].str.replace(r'[\s\u00A0]+', ' ', regex=True).str.strip()

# Option 2: The "Burn it all down" approach (removes all non-standard whitespace)
# This replaces any sequence of whitespace-like characters with a single space
df_clean['review'] = df_clean['review'].str.replace(r'\s+', ' ', regex=True, flags=re.UNICODE).str.strip()

# Verification
df_verify = df_clean['review'].head(10).apply(lambda x: f"|{x}|")
#print(df_verify)
df_clean.head(5)
```

```python
df_clean.to_csv('../../data/processed.csv')
print("cleaned data saved successfully!")
```
