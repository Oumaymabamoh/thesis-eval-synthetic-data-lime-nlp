```python
# import the packages
import pandas as pd
```

```python
# read the data by specifying the csv file
df = pd.read_csv('../../data/raw.csv')
# display the first few raws of the data
df.head()
```

```python
#generate a statistical summary of dataframe
df.describe()
```

```python
# get a concise su,,ary of DataFrame
df.info()
```

```python
# check the missing value
df.isna().sum()
```

```python
# Find rows with any missing value
rows_with_missing_data = df[df.isnull().any(axis=1)]

print("\nRows with missing values:")
print(rows_with_missing_data)
```

```python
# Check the total percentage of missing data in  model column
missing_percentage_model = df['model'].isnull().mean() * 100
print(f"Percentage missing in model: {missing_percentage_model}%")

# Check the total percentage of missing data in sentiment column
missing_percentage_senti = df['sentiment'].isnull().mean() * 100
print(f"Percentage missing in sentiment: {missing_percentage_senti}%")
```

```python

counts_senti = df['sentiment'].value_counts()
counts_model = df['model'].value_counts()
print(counts_senti)
print(counts_model)
```

```python
df_clean = df.dropna(subset=['model','sentiment'],how='any')
df_clean.head()
```
